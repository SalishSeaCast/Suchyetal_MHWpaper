## Add Daily Fraser River Data

```python
#dfFra=pd.read_csv('/ocean/eolson/MEOPAR/obs/ECRivers/Flow/FraserHopeDaily__Feb-8-2021_06_29_29AM.csv',
#                          skiprows=1)
df=pd.read_csv('/ocean/ksuchy/MOAD/analysis-karyn/notebooks/MHW/FraserRiver_Daily_ECCC_Sep2024.csv',skiprows=1)
```

```python
df.drop(df.loc[df.PARAM==2].index,inplace=True)
```

```python
df.rename(columns={'Value':'Flow'}, inplace=True)
```

```python
df2=df.loc[(df['YEAR'] >=2007)&(df['YEAR'] <= 2022)]
```

```python
df2
```

```python
df3 = df2.copy()

# Convert to full datetime
df3['Date'] = pd.to_datetime(df3['YEAR'] * 1000 + df3['DD'], format='%Y%j')

# Extract month
df3['Month'] = df3['Date'].dt.month
df3['YEAR'] = df3['Date'].dt.year  # just in case Year column isn't consistent

# Calculate monthly means
monthly_means = df3.groupby(['YEAR', 'Month'])['Flow'].mean().reset_index()
```

```python
monthly_means
```

```python
# Calculate climatology
climatology = monthly_means.groupby('Month')['Flow'].mean().rename('Climatology')
climatology_repeated = np.tile(climatology, 16)
# Calculate anomalies
monthly_anomalies = monthly_means.merge(climatology, on='Month')
monthly_anomalies['Anomaly'] = monthly_means['Flow'] - climatology_repeated
```
